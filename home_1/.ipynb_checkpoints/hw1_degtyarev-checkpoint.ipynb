Требуется написать функцию, которая на вход получает объект подключения к API и индекс, до какого объекта требуется запросить. Возвращать она будет pandas.DataFrame с полученными данными.

# Задание 0.Требуется написать функцию, которая на вход получает объект подключения к API и индекс, до какого объекта требуется запросить. Возвращать она будет pandas.DataFrame с полученными данными.

```python
def get_place_info(vkapi,idx=5000):
    query=[]
    list_result = [x for x in xrange(1, idx, 1000)]
    list_result.append(idx)
    for i in xrange(0,len(list_result)-1):
        list_from_l_to_r=range(list_result[i],list_result[i+1])
        time.sleep(0.5)
        query_temp=vkapi.places.getById(places=list_from_l_to_r)
        query.extend(query_temp)
    data=pd.DataFrame(query)
    return data
```

```python
data=get_place_info(vkapi,5000)
```

Замечание. Функция выгружает данные частями, так как невозможно выгрузить данные одним запросом (например если данных 5000), ограничение приблизительно в 2000.

# Задание 1. Провизуализируйте на карте все объекты которые находятся в Москве? Где больше объектов, в центре или на окраине?

```python
mos_center = [55.750475, 37.619020]
mos_min_lat, mos_max_lat = 55.564141, 55.908930
mos_min_lon, mos_max_lon = 37.344362, 37.879945
```

```python
data_moscow = data[data.latitude.between(mos_min_lat, mos_max_lat) &
                   data.longitude.between(mos_min_lon, mos_max_lon)]
```

```python
moscow = folium.Map(location=mos_center, zoom_start=11, control_scale=True)
```

```python
random_places = data_moscow[['latitude', 'longitude']]
```

```python
marker_cluster = MarkerCluster(locations=random_places[['latitude', 'longitude']].values)
```

```python
marker_cluster.add_to(moscow)
```

```python
moscow
```

```python
center_cor = np.array([[55.750417, 37.582659],[55.773840, 37.620425], [55.751772, 37.655959], [55.729692, 37.618022], [55.734148, 37.597422], [55.766969, 37.592101], [55.769582, 37.645659], [55.736182, 37.646517], [55.741799, 37.584376] ])
```

```python
def get_radius(center, random_cor):
    radius_list = []
    r=0
    for i in random_cor:
        x=(((i[0] - center[0])**2)+((i[1] - center[1]) ** 2))**1/2
        radius_list.append(x)
    return radius_list
```

```python
r = get_radius(mos_center, center_cor)
```

```python
avr = sum(r)/len(r)
```

```python
mask = (random_places.latitude-mos_center[0])**2+(random_places.longitude-mos_center[1])**2<=avr
```

```python
places_in_center=random_places[mask]
```

```python
moscow_only_center = folium.Map(location=mos_center, zoom_start=11, control_scale=True)
```

```python
marker_cluster = MarkerCluster(locations=places_in_center.values)
```

```python
marker_cluster.add_to(moscow_only_center)
```

```python
moscow_only_center
```

```python
places_not_in_center_count=len(random_places)-len(places_in_center)
```

```python
places_not_in_center_count
```

```python
len(places_in_center)
```

Замечание. Для того, чтобы выделить точки в центре, я взял точки на окраине садового кольца (массив center_cor), посчитал евклидово расстояние от центра до каждой из точек, и нашел среднее avr. Окружность отвечающая этому радиусу (приблизительно) и будет геометрическим местом всех точек, находящихся в центре.

Количество объектов в центре: 324 (внутри Садового кольца), количество объектов на окраине: 1347. Объектов в центре меньше.
